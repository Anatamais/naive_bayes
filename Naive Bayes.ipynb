```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import sklearn.preprocessing as skpre
%matplotlib inline
```

```python
data = pd.read_csv('credit-data.csv')
data.describe()
```

```python
data = data.drop("clientid", axis = 1)
data
```

```python
media = data['age'][data.age > 0].mean()
print(media)
data.loc[data.age < 0, 'age'] = media
data.loc[pd.isnull(data['age']), 'age'] = media
print(data)
```

```python
ss_income = skpre.StandardScaler()
ss_age = skpre.StandardScaler()
ss_loan = skpre.StandardScaler()
```

```python
data.income = ss_income.fit_transform(np.array(data['income']).reshape(-1, 1))
data.age = ss_age.fit_transform(np.array(data['age']).reshape(-1, 1))
data.loan = ss_loan.fit_transform(np.array(data['loan']).reshape(-1, 1))
```

```python
data_x = data.iloc[:, :3]
data_y = data.iloc[:, 3]
from sklearn.model_selection import train_test_split
data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x, data_y, test_size = 0.1)
```

```python
from sklearn.naive_bayes import GaussianNB
classificador = GaussianNB()
classificador.fit(data_x_train, data_y_train)
```

```python
cont_true, cont_false = 0, 0

for k in range(1700, 1999):
    resultado = classificador.predict([data.iloc[k, :3]])
    if resultado == data.iloc[k, 3]:
        cont_true += 1
    else:
        cont_false += 1
print("labels: ", classificador.classes_)
print("número de 0 e 1(label): ", classificador.class_count_)
print("probabilidade a priori de 0 e 1: ", classificador.class_prior_)
```

```python
print(cont_true)
print(cont_false)
```

```python
print("accuracy: ", cont_true/(cont_true + cont_false) * 100)
```
